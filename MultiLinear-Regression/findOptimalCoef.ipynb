```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = [1, 2, 5, 6]
W = [1, 1, 1, 1]
b = 2

y = b + sum((wi * xi) for wi, xi in zip(W, X))
y
```

```python
import random
def lin_reg(X, W, b):
    y = b + sum((wi * xi) for wi, xi in zip(W, X))
    return y

X1 = []
X2 = []
Y = []
for i in range(1000):
    x1 = random.uniform(-5, 5)
    x2 = random.uniform(-5, 5)
    X = [x1, x2]
    y = lin_reg(X, W, b)
    X1.append(x1)
    X2.append(x2)
    Y.append(y)

ax = plt.axes(projection="3d")
ax.view_init(30, 100)
ax.scatter3D(X1, X2, Y)
plt.show()
```
