```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def readData():
    file = open('boston', 'r')
    bostonHouse = []
    for x in file:
        bostonHouse += [x[:-1]]
    return bostonHouse

def aboutData(data):
    info = ''
    index = 0
    for i in data:
        if i == ' Variables in order:':
            break
        index += 1
        info += i
    return index, info
        
def aboutColumns(data, ind):
    data = data[ind + 1:]
    colInd = 0
    for i in data:
        if i == '':
            break
        colInd += 1
    data = [dt[1:] for dt in data[:colInd]]
    columns = {i.split(maxsplit=1)[0]:i.split(maxsplit=1)[1] for i in data}
    return colInd,columns

def getData(data, ind):
    data = data[ind + 1:]
    return [(data[i] + data[i + 1]).split() for i in range(0, len(data), 2)]

    
houseDb = readData()
index, info = aboutData(houseDb)
ind, columns = aboutColumns(houseDb, index)
col = list(columns.keys())
index += ind + 1
houseDb = getData(houseDb, index)
houseDb = pd.DataFrame(houseDb, columns=col)
houseDb = houseDb.apply(pd.to_numeric)
houseDb['CHAS'] = houseDb['CHAS'].astype('float64')
houseDb['RAD'] = houseDb['RAD'].astype('float64')
houseDb.head(20)
```

```python
X = houseDb[['RM', 'LSTAT']].values
Y = houseDb['MEDV'].values

model = linear_model.LinearRegression()
model.fit(X, Y)
print(f'Weight(w) : {model.coef_}')
print(f'Biais(b) : {model.intercept_}')
y_pred = model.predict(X)
y_pred[:20]
```
