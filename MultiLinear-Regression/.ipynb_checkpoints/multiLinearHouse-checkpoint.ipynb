```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def readData():
    file = open('boston', 'r')
    bostonHouse = []
    for x in file:
        bostonHouse += [x[:-1]]
    return bostonHouse

def aboutData(data):
    info = ''
    index = 0
    for i in data:
        if i == ' Variables in order:':
            break
        index += 1
        info += i
    return index, info
        
def aboutColumns(data, ind):
    data = data[ind + 1:]
    colInd = 0
    for i in data:
        if i == '':
            break
        colInd += 1
    data = [dt[1:] for dt in data[:colInd]]
    columns = {i.split(maxsplit=1)[0]:i.split(maxsplit=1)[1] for i in data}
    return colInd,columns

def getData(data, ind):
    data = data[ind + 1:]
    return [(data[i] + data[i + 1]).split() for i in range(0, len(data), 2)]

    
houseDb = readData()
index, info = aboutData(houseDb)
ind, columns = aboutColumns(houseDb, index)
col = list(columns.keys())
index += ind + 1
houseDb = getData(houseDb, index)
houseDb = pd.DataFrame(houseDb, columns=col)
houseDb = houseDb.apply(pd.to_numeric)
houseDb['CHAS'] = houseDb['CHAS'].astype('float64')
houseDb['RAD'] = houseDb['RAD'].astype('float64')
houseDb
```

```python
plt.plot(houseDb['RM'], houseDb['MEDV'], 'o', label='RM')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()
```

```python
plt.plot(houseDb['LSTAT'], houseDb['MEDV'], 'o', label='LSTAT')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.show()
```

```python
lstat = houseDb['LSTAT'].values
medv =  houseDb['MEDV'].values
rm = houseDb['RM'].values

def lin_reg(b, W, X):
    return (sum(i * j for i,j in zip(W, X)) + b)

b = 0
W = [1, 1]
Y = []
for i, j in zip(lstat, rm):
    X = [i, j]
    Y.append(lin_reg(b, W, X))

print(f'Value predict{Y[:10]}')
print(f'Value default{medv[:10]}')
```

```python
plt.plot(Y[:100], label='predictions')
plt.plot(medv[:100], label='prix medv')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y, medv)
mse
```
