```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
iRisData = pd.read_csv('iris.csv')
withoutSepal = iRisData.drop(['sepal_length', 'sepal_width'],axis=1)
withoutSepal
```

```python
plt.plot(withoutSepal['petal_length'], withoutSepal['petal_width'], 'o')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

```python
## i can manually ajust b and w to find the optimal line to predict my data
w = -0.4
b = 0.4
x = withoutSepal['petal_length']
y = w + (b * x)

plt.plot(withoutSepal['petal_length'], withoutSepal['petal_width'], 'o', label='withoutSepal')
plt.plot(x, y, label='predictions')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend()
plt.show()
```
